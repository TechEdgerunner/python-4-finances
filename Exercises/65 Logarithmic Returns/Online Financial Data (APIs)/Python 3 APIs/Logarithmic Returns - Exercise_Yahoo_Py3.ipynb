## Logarithmic Returns

```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
MSFT = yf.download('MSFT', start='2000-1-1')
MSFT
```

### Log Returns

$$
ln(\frac{P_t}{P_{t-1}})
$$

Calculate the Log returns for Microsoft.

```python
MSFT['log_return'] = np.log(MSFT['Adj Close'] / MSFT['Adj Close'].shift(1))
print(MSFT['log_return'])
```

Plot the results on a graph.

```python
MSFT['log_return'].plot(figsize=(16, 9))
plt.show()
```

Estimate the daily and the annual mean of the obtained log returns.

```python
daily_return = MSFT['log_return'].mean()
```

```python
annual_return = MSFT['log_return'].mean() * 250
```

Print the result in a presentable form.

```python
print('Annual return: ',round(annual_return*100, 2), '%')
print('Daily return: ',round(daily_return*100, 2), '%')
```
